# Data Preparation

```python
# remove gust and peak_wind_drct from the dataset as there are alot of missing values
dataset.drop(['gust', 'peak_wind_drct'], axis=1, inplace=True)
dataset.head()
```

```python
# print names of all the columns
dataset.columns
```

```python
# rename ' station' to 'station'
dataset.rename(columns={' station': 'station'}, inplace=True)
```

```python
# print unique values of station
dataset['station'].unique()
```

```python
# remove station column from the dataset as it is not needed
dataset.drop(['station'], axis=1, inplace=True)
```

```python
dataset.tail()
```

```python
# remove all NaN values from the dataset
dataset.dropna(inplace=True)
```

```python
dataset.tail()
```

```python
dataset.dtypes
```

```python
# remove date_time column from the dataset as it is not needed
dataset.drop(['date_time'], axis=1, inplace=True)
# drop full_date column as it is not needed
dataset.drop(['full_date'], axis=1, inplace=True)
```

```python
dataset.head()
```

```python
dataset.tail()
```

```python
# create train and test datasets
# make training set all values upto 2018
train = dataset[dataset['year'] < 2018]
# make test set all values from 2018 onwards
test = dataset[dataset['year'] >= 2018]
```

```python
# create x train and y train datasets
# the predictor is the year, month, day, hour and minute columns
x_train = train[['year', 'month', 'day', 'hour', 'minute']]
# the target is every other column except year, month, day, hour and minute columns
y_train = train.drop(['year', 'month', 'day', 'hour', 'minute'], axis=1)

# create x test and y test datasets
x_test = test[['year', 'month', 'day', 'hour', 'minute']]
y_test = test.drop(['year', 'month', 'day', 'hour', 'minute'], axis=1)
```

### Reasons for changes in the data
Firstly I decided to remove the gust and peak_wind_drct as there are alot of missing values. I also decided to remove the station column because there was only one unique station in the dataset which was NZAA. Thirdly I decided to split the date column into year, month, day, hour and minute so that I could easily pass the on the date values into the model. Finally, when creating the train and test datasets I decided to split the data so that all the training data was up to the year 2018 and the test data was everything after.

# Data Modelling

```python
# An exhaustive list of models can be found in sci-kit learn in this link: https://scikit-learn.org/stable/supervised_learning.html
# The url takes you to supervised learning which is what you will be dealing with for this phase.
```

```python
# import neighbors from sklearn
from sklearn.neighbors import KNeighborsRegressor
```

```python
# train the model on the train dataset
neigh = KNeighborsRegressor()
neigh.fit(x_train, y_train)
```

```python
# test the model on the test dataset and print the accuracy
print("KNeighborsRegressor Accuracy: " + str(neigh.score(x_test, y_test)))
```

### My Prediction results will print in the order tmpc, dwpc, relh, sknt

```python
# predict the values for the year = 2021, month = 1, day = 29, hour = 0, minute = 0
neigh.predict([[2021, 1, 29, 21, 30]])
```
